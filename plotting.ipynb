```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import date
```

```python
# generate the data
t = np.arange(0, 10, 0.01)

x = np.sin(t)  # position
y = np.sin(2*t)

vx = np.cos(t) # velocity
vy = 2*np.cos(2*t)
```

```python
fig, axs = plt.subplots(nrows= 2,
                        ncols= 1,
                        sharex=True)
axs[0].plot(t,
            np.column_stack([x,y]),
            linewidth= 2)
axs[1].plot(t,
            np.column_stack([vx, vy]),
            linewidth=2)

axs[0].set_ylabel('Position [m]')
axs[1].set_xlabel('Time [sec]')
axs[1].set_ylabel('Velocity [m/s]')

axs[0].set_xlim([0,10])

fig.text(.8, 0.02, date.isoformat(date.today()))

for ax in axs:
    lines = ax.get_lines()
    lines[0].set_linestyle('dashed')


axs[0].legend(['x', 'y'], loc='best')

out_file = 'simple_figure.jpg'
plt.savefig(out_file, dpi=300)

fig.set_facecolor('white')
axs[0].set_facecolor('lightgreen')
```
